```python
# 7.
from functools import wraps
import time


def timeit(func):
    @wraps(func)
    def timeit_wrapper(*args, **kwargs):
        start_time = time.perf_counter()
        result = func(*args, **kwargs)
        end_time = time.perf_counter()
        total_time = end_time - start_time
        # first item in the args, ie `args[0]` is `self`
        print(f'Function {func.__name__} Took {total_time:.4f} seconds')
        return result
    return timeit_wrapper


class Calculator:
    @timeit
    def calculate_something(self, num):
        """
        an example function that returns sum of all numbers up to the square of num
        """
        total = sum((x for x in range(0, num**2)))
        return total

    def __repr__(self):
        return f'calc_object:{id(self)}'


if __name__ == '__main__':
    calc = Calculator()
    print(calc.calculate_something(10))
    
```

```python
# 9, 10, 11, 12
import time
class Robot:
    def __init__(self, current_position=[0.0,0.0], current_direction=[1.0, 0.0], current_time=0.0, speed=0.5):
        self.current_position=current_position
        self.current_direction=current_direction
        self.current_time=current_time
        self.speed=speed
    def __repr__(self):
        return "current position="+str(self.current_position)+"current_direction="+str(self.current_direction)+"current_time="+str(self.current_time)
    def wait(self, Time):
        time.sleep(Time)
        self.current_time+=Time

    def move_forward(self, Time):
        self.current_position[0]+=self.current_direction[0]*self.speed*Time
        self.current_position[1]+=self.current_direction[1]*self.speed*Time
        self.current_time+=Time

    def rotate_right(self):
        if self.current_direction==[1, 0]:
            self.current_direction=[0, -1]
        elif self.current_direction==[0, -1]:
            self.current_direction=[-1, 0]
        elif self.current_direction==[-1, 0]:
            self.current_direction=[0, 1]
        else: 
            self.current_direction=[1, 0]
        
    def rotate_left(self):
        if self.current_direction==[1, 0]:
            self.current_direction=[0, 1]
        elif self.current_direction==[0, 1]:
            self.current_direction=[-1, 0]
        elif self.current_direction==[-1, 0]:
            self.current_direction=[0, -1]
        else: 
            self.current_direction=[1, 0]  
    def move_square(self, side_length):
        delta_time=side_length/self.speed
        for i in range(4):
            self.move_forward(delta_time)
            self.rotate_right()
            print(repr(self))

        
        
        



```
